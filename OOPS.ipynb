```
from datetime import date
from dateutil.relativedelta import relativedelta
class person:
  def __init__(self):
    self.name=None
    self.__age=None
    self.phno=None
  
  @property
  def details(self):
    return "Name:{}\nAge:{}\nPhNo:{}".format(self.name,self.__age,self.phno)
  
  @details.setter
  def details(self,det):
    self.name=det[0]
    da=list(map(int,det[1].split('-')))
    self.__age=date(da[2],da[1],da[0])
    self.phno=det[2]
    
    
  def detail(self,name,age,phno):
    self.name=name
    self.__age=age
    self.phno=phno
    
  @property
  def yob(self):
    year=date.today()
    return str(relativedelta(year,self.__age).years)
  
  def __str__(self):
    return self.name
    
```

```
class role(person):
  def __init__(self,name,age,phno,rol):
    self.rol=rol
    d=list(map(int,age.split('-')))
    dat=date(d[2],d[1],d[0])
    self.detail(name,dat,phno)
    
  @property
  def show(self):
    return "Name:{}\nAge:{}\nPhNo:{}".format(self.name,self.phno,self.rol)
  
  def __str__(self):
    return self.name
    
```

```
x=role('Harsh Goyal','17-07-1997','9212','Manager')
print(x)
print(x.show)
```

```
print(x.name)
print(x.phno)
print(x.yob)
```
