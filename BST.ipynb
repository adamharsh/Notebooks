```
class Node:
  def __init__(self,data):
    self.data=data
    self.left=None
    self.right=None
```

```
class BST:
  def __init__(self):
    self.root=None
  
  
  def add(self,data,root):
    if root==None:
      self.root=Node(data)
    elif data<root.data:
      if root.left==None:
        root.left=Node(data)
      else:
        self.add(data,root.left)
    elif data>root.data:
      if root.right==None:
        root.right=Node(data)
      else:
        self.add(data,root.right)
      
  def add2(self,data):
    if self.root==None:
      self.root=Node(data)
    else:
      temp=self.root
      while(1):
        if data<temp.data:
          if temp.left==None:
            temp.left=Node(data)
            break
          else:
            temp=temp.left
        elif data>temp.data:
          if temp.right==None:
            temp.right=Node(data)
            break
          else:
            temp=temp.right
    
  
  
  def inorder(self,root):
    if root!=None:
      
      self.inorder(root.left)
      print(root.data)
      self.inorder(root.right)
      
  def preorder(self,root):
    if root!=None:
      print(root.data)
      self.preorder(root.left)
      self.preorder(root.right)
      
  def postorder(self,root):
    if root!=None:
      self.postorder(root.left)
      self.postorder(root.right)
      print(root.data)
      
#   def delete1(self,data):
#     temp=self.root
#     while(1):
#       if self.root==None:
#       print("empty")
#       break
#       elif data==temp:
#         if temp.left !=None:
#           temp=temp.left
#           while temp.right!=None:
#             temp=temp.right
#           self.root=temp.data
#           temp=None
#         elif temp.right!=None:
#           temp=temp.right
#           while temp.left!=None:
#             temp=temp.left
#           self.root=temp
#           temp=None
#         else:
#           root=None
  
  
  def __find(self,temp):
    if temp.left==temp.right==None:
      return None
    else:
      if temp.left!=None:
        x=temp.left
        if temp.left.right==None:
          data=temp.left.data
          temp.left=None
          return data
        while(1):
          if x.right.right==None:
            data=x.right.data
            x.right=None
            return data
          else:
            x=x.right
      elif temp.right!=None:
        x=temp.right
        if temp.right.left==None:
          data=temp.right.data
          temp.right=None
          return data
        while(1):
          if x.left.left==None:
            data=x.left.data
            x.left=None
            return data
          else:
            x=x.left
                  
          
  
  def delete2(self,data):
    if self.root==None:
      print("Empty Tree")
    else:
      temp=self.root
      prev_temp=temp
      found=0
      while(1):
        if temp!=None and temp.data==data:
          found=1
          break
        elif temp!=None and data<temp.data:
          prev_temp=temp
          temp=temp.left
        elif temp!=None and data>temp.data:
          prev_temp=temp
          temp=temp.right
        else:
          break
          
      if found!=1:
        print("Node Not found")
      else:
        data=self.__find(temp)
        if data!=None:
          temp.data=data
        else:
          if prev_temp.data>temp.data:
            prev_temp.left=None
          elif prev_temp.data<temp.data:
            prev_temp.right=None
          else:
            self.root=None
          
```
