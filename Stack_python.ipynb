```
class node:
  def __init__(self,data,nex):
    self.data=data
    self.nex=nex
    
    
class stack:
  def __init__(self):
    self.top=None
    
    
  def push(self,data):
    if self.top is None:
      self.top=node(data,None)
    else:
      temp=node(data,self.top)
      self.top=temp
  
  
  def pop(self):
    print(self.top.data)
    self.top=self.top.nex
    
  
  def show(self):
    temp=self.top
    while(temp is not None):
      print(temp.data)
      temp=temp.nex

      
  def show_top(self):
    if self.top is None:
      print(None)
    else:
      print(self.top.data)
  
  
  def __str__(self):
    if self.top is None:
      y=None
    else:
      y=self.top.data
    return str(y)
```

```
x=stack()
```

```
x.push(1)
```
