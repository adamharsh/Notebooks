```
if(n%6!=0):
  n=6-n%6
  zer="0"
  zer=zer*n
  binary=zer+x
else:
  binary=x
binary=list(binary)
res=''
i=0
sum_b=0
while i < len(binary):
  x=5-i%6
  sum_b+=int(binary[i])*(2**x)
  if(x==0):
    res+=data_dic[sum_b]
    sum_b=0  
  i+=1
print(res)
```
