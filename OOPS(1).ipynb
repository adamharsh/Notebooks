**CONSTRUCTOR IN PYTHON**

```
class Employee:
  def __init__(self,name):
    self.name=name
  def __str__(self):
    return "Welcome {}".format(self.name)
```

```
e1=Employee("Harsh")
print(e1)
```

**CLASS METHOD**

```
class Employee:
  @classmethod
  def from_string(cls,detail):
    data=detail.split("-")
    print(data)
    cls.name=data[0]
    cls.phno=data[1]
    return cls
```

```
emp1=Employee.from_string("Harsh-09876543")
print(emp1.phno)
```

**STATIC METHOD**

```
class Employee:
  @staticmethod
  def is_good(rate):
    if rate>=3:
      return "Good"
    else:
      return "Not Good"
```

```
Employee.is_good(4)
```

**INHERITANCE**

```
class Employee:
  rate=1.9054
  def prin(self):
    print(self.rate)
```

```
class Developer(Employee):
  def __str__(self):
    super().prin()
    return str(super().rate)
```

```
a=Developer()
print(a)
```
